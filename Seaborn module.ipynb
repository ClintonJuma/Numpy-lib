# Seaborn module

Used for visualizing distribution/
Distplot-Distribution Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5], hist = False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot([0, 1, 2, 3, 4, 5])
plt.show()
```
